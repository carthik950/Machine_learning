```python
# Since the values of the features are in different orders of magnitude, lets use standardscaler to scale the data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler_X = scaler.fit_transform(df)
```

```python
scaler_X
```

```python
# pd.DataFrame(scaler_X, columns = df.columns)
```

```python
from sklearn.cluster import DBSCAN
```

```python
# we need to check the outlier percentage 
```

```python
outlier_percentage = []

for eps in np.linspace(0.001,3,50):
    dbscan = DBSCAN(eps = eps, min_samples = 2* scaler_X.shape[1])
    dbscan.fit(scaler_X)
    
    perc_outlier = 100*np.sum(dbscan.labels_ ==-1)/ len(dbscan.labels_)
    outlier_percentage.append(perc_outlier)
```

```python
# scaler_X.shape[1]
```

```python
# Create a line plot of the percentage of outlier points vs the epsilon value choice
```

```python
sns.lineplot(x = np.linspace(0.001,3,50), y = outlier_percentage )
```

```python
# DBSCAN with chosen Epsilon
```

```python
dbscan = DBSCAN(eps = 2, min_samples = 2* scaler_X.shape[1])
```

```python
dbscan.fit(scaler_X)
```

```python
dbscan.labels_ # outliers # -1 is outlier
```

```python
sns.scatterplot(data = df, x = 'Grocery', y = 'Milk', hue = dbscan.labels_,palette = 'Set1' )
```

```python
# Create a scatterplot of milk vs Detergents paper colored by the lables
```

```python
sns.scatterplot(data = df, x ='Detergents_Paper', y = 'Milk', hue = dbscan.labels_, palette = 'Set1')
```
