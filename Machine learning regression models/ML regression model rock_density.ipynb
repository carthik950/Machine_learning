# POLYNOMIAL REGRESSION

```python
def run_model(model,X_train,y_train,X_test,y_test):
    # Fit model training
    model.fit(X_train,y_train)
        
    # Get performance metrics
    preds = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test,preds))
    mae = mean_absolute_error(y_test,preds)
    print('MAE:', mae)
    print('RMSE:', rmse)
    
    # Plot results model signal range
    signal_range = np.arange(0,100)
    signal_pred = model.predict(signal_range.reshape(-1,1))
    
    plt.figure(figsize=(12,8))
    sns.scatterplot(x ='Signal', y ='Density', data = df, color = 'black')
    plt.plot(signal_range,signal_pred)
    
    
```

```python
# pipeline
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe  = make_pipeline(PolynomialFeatures(degree=6), LinearRegression()) # Transform x_train & y_train into polynomial features
```

```python
model = LinearRegression()
run_model(model,X_train,y_train,X_test,y_test)
```

```python
run_model(pipe,X_train,y_train,X_test,y_test)
```

# KNN Model

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
k_values = [1,5,10] # if we increase k value it will be based

for n in k_values:
    model = KNeighborsRegressor(n_neighbors=n)
    run_model(model,X_train,y_train,X_test,y_test)
```

# Decision Tree model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
run_model(model,X_train,y_train,X_test,y_test)
```

# SVM MODEL

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
model_svr = SVR()
param_grid = {'C': [0.01,0.1,1,5,10,100,1000], 'gamma': ['auto', 'scale']}

grid = GridSearchCV(model_svr,param_grid)
```

```python

run_model(grid,X_train,y_train,X_test,y_test)
```

# Random forest model

```python
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor(n_estimators=10)
```

```python
run_model(model_rf,X_train,y_train,X_test,y_test)
```

# Gradient Boosting & Adaboost model

```python
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
```

```python
model_gb = GradientBoostingRegressor()
```

```python
run_model(model_gb,X_train,y_train,X_test,y_test)
```

```python
model_ada = AdaBoostRegressor()
run_model(model_ada,X_train,y_train,X_test,y_test)
```
